```python
# Make a Graph class with Matrix Representation
class Graph:
    def __init__(self, root=None, weighted=None, directed=None, h_flag=None, root_h_val=None):
        # Matrix representation instead of adjacency list
        self.__vertices = []
        self.__vertex_to_index = {}
        self.__index_to_vertex = {}
        self.__matrix = []
        
        if root:
            self.__vertices.append(root)
            self.__vertex_to_index[root] = 0
            self.__index_to_vertex[0] = root
            self.__matrix = [[None]]
        else:
            self.__vertices.append("root")
            self.__vertex_to_index["root"] = 0
            self.__index_to_vertex[0] = "root"
            self.__matrix = [[None]]

        self.__h_flag = h_flag
        if h_flag == True:
            self.__h_flag = h_flag
            if root_h_val:
                self.__h_map = {self.__vertices[0]: root_h_val}
            else:
                raise Exception("There should be an heuristic value for the root")
        elif self.__h_flag != None:
            raise Exception("h_flag parameter should be True or None")

        self.__weighted = weighted
        if weighted == True:
            self.__weighted = weighted
        elif self.__weighted != None:
            raise Exception("weighted parameter should be True or None")

        self.__directed = directed
        if directed == True:
            self.__directed = directed
        elif self.__directed != None:
            raise Exception("directed parameter should be True or None")

        print(f"Graph is created with root name: {self.__vertices[0]}")

    #________________________________________________________________________________________________________
    def add_vertex(self, vertex, h_val=None):
        if vertex not in self.__vertices:
            # Add vertex to list
            self.__vertices.append(vertex)
            idx = len(self.__vertices) - 1
            self.__vertex_to_index[vertex] = idx
            self.__index_to_vertex[idx] = vertex
            
            # Expand matrix: add new row and column
            for row in self.__matrix:
                row.append(None)
            self.__matrix.append([None] * len(self.__vertices))
            
            print("Vertex is added successfully !")
            if self.__h_flag:
                if h_val is not None:
                    self.__h_map[vertex] = h_val
                else:
                    raise Exception("There should be an heuristic value for the vertex")
        else:
            print("This vertex does exist in the graph already !")
    
    #________________________________________________________________________________________________________
    def add_edge(self, vertex_1, vertex_2, weight=None):
        if vertex_1 in self.__vertices and vertex_2 in self.__vertices:
            idx1 = self.__vertex_to_index[vertex_1]
            idx2 = self.__vertex_to_index[vertex_2]
            
            if not self.__directed:
                if not self.__weighted:
                    self.__matrix[idx1][idx2] = 1
                    self.__matrix[idx2][idx1] = 1
                else:
                    if weight != None:
                        self.__matrix[idx1][idx2] = weight
                        self.__matrix[idx2][idx1] = weight
                    else:
                        raise Exception("It should be a weight for the edge")
            else:
                if not self.__weighted:
                    self.__matrix[idx1][idx2] = 1
                else:
                    if weight != None:
                        self.__matrix[idx1][idx2] = weight
                    else:
                        raise Exception("It should be a weight for the edge")
        else:
            raise Exception("It should be both vertices exist in the graph !")

        print(f"Edge added between {vertex_1} and {vertex_2}")

    #________________________________________________________________________________________________________
    def get_graph(self):
        try:
            return self.__matrix, self.__vertices, self.__h_map
        except:
            return self.__matrix, self.__vertices
    
    #________________________________________________________________________________________________________
    def get_adjacent_nodes(self, vertex):
        """Get adjacent nodes from matrix representation"""
        if vertex not in self.__vertex_to_index:
            return []
        
        idx = self.__vertex_to_index[vertex]
        adjacent = []
        
        for i, weight in enumerate(self.__matrix[idx]):
            if weight is not None:
                adj_vertex = self.__index_to_vertex[i]
                if self.__weighted:
                    adjacent.append((adj_vertex, weight))
                else:
                    adjacent.append(adj_vertex)
        
        return adjacent
    
    #________________________________________________________________________________________________________
    def basic_search(self, goal, start=None):
        """Interface method for search - to be overridden by child classes"""
        raise NotImplementedError("This method should be implemented by child classes")
    
    #________________________________________________________________________________________________________
    def _heuristic_cost(self, path):
        """Helper method to get heuristic cost of a path"""
        return self.__h_map[path[-1]], path[-1]
```

```python
# Create Best_first class that inherits from Graph
class Best_first(Graph):
    def basic_search(self, goal, start=None):
        """Best-First Search implementation using polymorphism"""
        if self._Graph__h_flag is None:
            raise Exception("Best First Search only works with Graphs with Heuristic values !")
        
        if start != None:
            lst = [[start]]
        else:
            lst = [["root"]]

        visited = []
        while lst:
            # Sort all elements based on heuristic cost and alphabetical order
            lst.sort(key=self._heuristic_cost)
            # basic search
            current_path = lst.pop(0) # FIFO
            current_node = current_path[-1]
            
            if current_node in visited:
                continue
            visited.append(current_node)
            
            if current_node == goal:
                return current_path, visited
            else:
                adjacent_nodes = self.get_adjacent_nodes(current_node)
                if self._Graph__weighted:
                    for node, _ in adjacent_nodes:
                        new_path = current_path.copy()
                        new_path.append(node)
                        lst.append(new_path)
                else:
                    for node in adjacent_nodes:
                        new_path = current_path.copy()
                        new_path.append(node)
                        lst.append(new_path)

        raise Exception("Search Failed !")

# Create Greedy class that inherits from Graph
class Greedy(Graph):
    def basic_search(self, goal, start=None):
        """Greedy Search implementation using polymorphism"""
        if self._Graph__h_flag is None:
            raise Exception("Greedy Search only works with Graphs with Heuristic values !")
        
        if start != None:
            lst = [[start]]
        else:
            lst = [["root"]]
    
        visited = []
        while lst:
            # Sort all elements based on heuristic cost and alphabetical order
            lst.sort(key=self._heuristic_cost)
            # basic search
            current_path = lst.pop(0) # FIFO
            lst.clear() # Clear the list - this is the key difference for Greedy
            current_node = current_path[-1]
            
            if current_node in visited:
                continue
            visited.append(current_node)
            
            if current_node == goal:
                return current_path, visited
            else:
                adjacent_nodes = self.get_adjacent_nodes(current_node)
                if self._Graph__weighted:
                    for node, _ in adjacent_nodes:
                        new_path = current_path.copy()
                        new_path.append(node)
                        lst.append(new_path)
                else:
                    for node in adjacent_nodes:
                        new_path = current_path.copy()
                        new_path.append(node)
                        lst.append(new_path)
    
        return "Search Failed"
```
